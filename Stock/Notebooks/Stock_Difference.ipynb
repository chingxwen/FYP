```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Jurgen\Desktop\FYP\Stock\2018-07-01.2019-10-31.005930.ks.csv")
df
```

```python
close = df[["date", "Close_USD"]] 
close = close.rename(columns={"date": "Date"})
close["Date"] = pd.to_datetime(close["Date"], format = "%d/%m/%Y").dt.date
close = close.set_index("Date")
```

```python
close
```

```python
diff = [0]
for i in range(len(close) - 1):
    diff.append(close["Close_USD"][i+1] - close["Close_USD"][i])
print(len(diff))
```

```python
close["Stock_Difference"] = diff
```

```python
avg = close["Stock_Difference"].sum()/len(close["Stock_Difference"])
avg
```

```python
close["Stock_Difference"] = close["Stock_Difference"].replace(0, avg)
```

```python
close.to_csv(r"StockDifference.csv")
```
