```python
# Important imports

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import numpy as np
import os
```

```python
# Read freshly downloaded file
data = pd.read_csv("samsungtech15-19.csv")

# Remove rows with null values
data['High'] = data['High'].replace('null', np.nan)
dg = data.dropna(axis=0, subset=['High'])

#renaming columns
dg.rename(columns={'Adj Close':'adj_close', 'Date':'date', 'Volume': 'volume', 'Close':'close', 'Low': 'low', 'High': 'high', 'Low': 'low', 'Open':'open'}, inplace = True)

# Write to a new csv
#pd.DataFrame.from_dict(data=dg, orient='columns').to_csv("cleanedsamsungtech15-19.csv")

# Change data to a dataframe
df = pd.DataFrame(dg)
```

```python
# Insert new column data into new column, stating the stockname - so later can combine multiple files
Stockname = []
filename = "samsungtech15-19.csv"

for i in range (len(df.high)):
    Stockname.append(os.path.splitext(filename)[0])
    
# Add a column and column name
df.insert(7, "Stockname", Stockname , True)


print(Stockname)
print(df)
```

```python
pd.DataFrame.from_dict(data=df, orient='columns').to_csv("cleanedsamsungtesting.csv")
```
